```r
crdt_crd<-read.csv("D:\\R_TXT_MINE_DATA\\creditcard.csv",header = TRUE,
                   stringsAsFactors = FALSE,
                   strip.white = TRUE ,
                   sep = ',')

#checking overall dataset and coverting 'class' as factor variables 
str(crdt_crd)
crdt_crd$Class = as.factor(crdt_crd$Class)
str(crdt_crd$Class)
```

```r
#install.packages("partykit")
library(party)
library(partykit)
cc_dsn_tree <- ctree(Class ~V14+V10+V4+V16, data=crdt_crd)
cc_dsn_tree 
```

```r
plot(as.simpleparty(cc_dsn_tree ), type="simple")
```

```r
library(rpart)
library(rattle)

cc_dec_tree <- rpart(Class ~V14+V10+V4+V16, data=crdt_crd, method="class")
fancyRpartPlot(cc_dec_tree)
```
